In Python, the terms "mutable" and "immutable" refer to whether an object’s state can be modified after it has been created.
Here’s a comparison of mutable and immutable objects, along with examples to illustrate each concept.

Mutable Objects
Definition: Mutable objects are those whose state or contents can be changed after they are created.

Examples: Lists, dictionaries, sets.

Immutable Objects
Definition: Immutable objects are those whose state or contents cannot be changed after they are created.

Examples: Tuples, strings, integers, floats.

Characteristics:

Any operation that appears to modify an immutable object will actually create a new object.
Changes to an immutable object result in a new object being created, not a modification of the existing object

Basic Structure
python
Copy code
if condition:
    # code to execute if condition is True
elif another_condition:
    # code to execute if the previous condition is False and this one is True
else:
    # code to execute if all previous conditions are False
Examples
1. Basic if Statement
The if statement executes a block of code if its condition evaluates to True.

```python
In Python, loops are used to repeatedly execute a block of code while a condition is true or for a specified number of iterations. Python supports several types of loops, each serving different purposes:

1. for Loop
The for loop in Python is used to iterate over a sequence (like a list, tuple, string, or range) or other iterable objects. It’s useful when you know the number of iterations in advance.

Syntax:

python
Copy code
for variable in iterable:
    # code to execute
Examples:

Iterating Over a List:

python
Copy code
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
Output:

Copy code
apple
banana
cherry
Using range() for Repeated Iterations:

python
Copy code
for i in range(5):
    print(i)
Output:

Copy code
0
1
2
3
4
Here, range(5) generates a sequence of numbers from 0 to 4.

Iterating Over a String:

python
Copy code
for char in "hello":
    print(char)
Output:

Copy code
h
e
l
l
o
2. while Loop
The while loop repeatedly executes a block of code as long as a specified condition is True. It’s useful when you don’t know the number of iterations beforehand and want to continue looping until a condition changes.

Syntax:

python
Copy code
while condition:
    # code to execute
Examples:

Basic while Loop:

python
Copy code
count = 0
while count < 5:
    print(count)
    count += 1
Output:

Copy code
0
1
2
3
4
Here, the loop continues as long as count is less than 5. The value of count is incremented in each iteration.

Using while Loop with User Input:

python
Copy code
number = 0
while number != 5:
    number = int(input("Enter a number (5 to stop): "))
    print(f"You entered: {number}")
This loop will continue to prompt the user for input until they enter 5.

3. break Statement
The break statement is used to exit a loop prematurely, regardless of the loop’s condition.

Examples:

Using break in a for Loop:

python
Copy code
for i in range(10):
    if i == 5:
        break
    print(i)
Output:

Copy code
0
1
2
3
4
The loop exits when i equals 5, so the numbers 0 through 4 are printed.

Using break in a while Loop:

python
Copy code
count = 0
while True:
    if count == 3:
        break
    print(count)
    count += 1
Output:

Copy code
0
1
2
The while True loop runs indefinitely until the break statement is encountered.

4. continue Statement
The continue statement skips the rest of the code inside the loop for the current iteration and proceeds to the next iteration.

Examples:

Using continue in a for Loop:

python
Copy code
for i in range(5):
    if i == 2:
        continue
    print(i)
Output:

Copy code
0
1
3
4
When i equals 2, the continue statement skips the print statement and proceeds with the next iteration.

Using continue in a while Loop:

python
Copy code
count = 0
while count < 5:
    count += 1
    if count == 3:
        continue
    print(count)
Output:

Copy code
1
2
4
5
When count equals 3, the continue statement skips the print statement and continues to the next iteration.

5. Nested Loops
You can nest loops inside other loops to handle more complex scenarios.

Examples:

Nested for Loops:

python
Copy code
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
Output:

css
Copy code
i=0, j=0
i=0, j=1
i=1, j=0
i=1, j=1
i=2, j=0
i=2, j=1
Nested while Loops:

python
Copy code
i = 0
while i < 3:
    j = 0
    while j < 2:
        print(f"i={i}, j={j}")
        j += 1
    i += 1
Output:

css
Copy code
i=0, j=0
i=0, j=1
i=1, j=0
i=1, j=1
i=2, j=0
i=2, j=1
```
